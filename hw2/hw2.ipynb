```python
import numpy as np
```

```python
FEATURES_FILE = 'texts_train.txt'
LABELS_FILE = 'scores_train.txt'
```

```python
import re

regex = re.compile('[^а-яА-Я]')
#First parameter is the replacement, second parameter is your input string
regex.sub('', 'аь3ф*ЙЙ')
#Out: 'abdE'
```

```python
import operator
```

```python
dic = {}
dic2 = {}
dic3 = {}
lines = []
lines_2 = []
lines_3 = []
labels = []
with open(FEATURES_FILE, 'r') as fin:
    for line in fin:
        line_words = []
        line_2 = []
        line_3 = []
        prev_word = None
        prev_prev_word = None
        for word in line.split(' '):
            pure_word = regex.sub('', word).replace('ё', 'e').lower()
            if len(pure_word) <= 3:
                continue
            line_words.append(pure_word)
            if len(pure_word) > 3:
                if pure_word in dic.keys():
                    dic[pure_word] += 1
                else:
                    dic[pure_word] = 1
            if prev_word and prev_prev_word:
                line_3.append((prev_prev_word, prev_word, pure_word))
                if (prev_prev_word, prev_word, pure_word) in dic3.keys():
                    dic3[(prev_prev_word, prev_word, pure_word)] += 1
                else:
                    dic3[(prev_prev_word, prev_word, pure_word)] = 1
                prev_prev_word = prev_word
                prev_word = pure_word
            else:
                if prev_word:
                    if (prev_word, pure_word) in dic2.keys():
                        dic2[(prev_word, pure_word)] += 1
                    else:
                        dic2[(prev_word, pure_word)] = 1
                    line_2.append((prev_word, pure_word))
                    prev_prev_word = prev_word
                    prev_word = pure_word
                else:
                    prev_word = pure_word
        lines.append(line_words)
        lines_2.append(line_2)
        lines_3.append(line_3)
sorted_dict = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
sorted_dict2 = sorted(dic2.items(), key=operator.itemgetter(1), reverse=True)
sorted_dict3 = sorted(dic3.items(), key=operator.itemgetter(1), reverse=True)
words = [x[0] for x in sorted_dict][:1000]
words2 = [x[0] for x in sorted_dict2][:1000]
words3 = [x[0] for x in sorted_dict3][:1000]
word_to_index = {}
for i in range(len(words)):
    word_to_index[words[i]] = i

               
word_to_index2 = {}
for i in range(len(words2)):
    word_to_index[words2[i]] = i       
    
word_to_index3 = {}
for i in range(len(words3)):
    word_to_index[words3[i]] = i            
```

```python
with open(LABELS_FILE, 'r') as fin:
    for line in fin:
        labels.append(int(line))
```

```python
word_to_positive = {}
word_to_negative = {}
positive_number = 0
negative_number = 0
pos_threshold = 6
neg_threshold = 5
word_line_to_cnt = {}
```

```python
for i in range(len(lines)):
    if labels[i] >= pos_threshold:
        positive_number += 1
    if labels[i] <= neg_threshold:
        negative_number += 1
    
    word_to_cnt = {}
    for word in lines[i]:
        if word in word_to_cnt.keys():
            word_to_cnt[word] += 1
        else:
            word_to_cnt[word] = 1
        if labels[i] >= pos_threshold:
            if word in word_to_positive.keys():
                word_to_positive[word] += 1
            else:
                word_to_positive[word] = 1
        if labels[i] <= neg_threshold:
            if word in word_to_negative.keys():
                word_to_negative[word] += 1
            else:
                word_to_negative[word] = 1
    for word in word_to_cnt.keys():
        word_line_to_cnt[(word, i)] = word_to_cnt[word]
```

```python
for word in word_to_index.keys():
    if word not in word_to_positive.keys():
        word_to_positive[word] = 0.00000000000000000001
    if word not in word_to_negative.keys():
        word_to_negative[word] = 0.00000000000000000001    
```

```python
word2_to_positive = {}
word2_to_negative = {}
word2_line_to_cnt = {}
```

```python
for i in range(len(lines)):
    word_to_cnt = {}
    for word2 in lines_2[i]:
        if word2 in word_to_cnt.keys():
            word_to_cnt[word2] += 1
        else:
            word_to_cnt[word2] = 1
        if labels[i] >= pos_threshold:
            if word2 in word2_to_positive.keys():
                word2_to_positive[word2] += 1
            else:
                word2_to_positive[word2] = 1
        if labels[i] <= neg_threshold:
            if word2 in word2_to_negative.keys():
                word2_to_negative[word2] += 1
            else:
                word2_to_negative[word2] = 1
    
    for word in word_to_cnt.keys():
        word2_line_to_cnt[(word, i)] = word_to_cnt[word]
```

```python
word3_to_positive = {}
word3_to_negative = {}
word3_line_to_cnt = {}
```

```python
for i in range(len(lines)):
    word_to_cnt = {}
    for word3 in lines_3[i]:
        if word3 in word_to_cnt.keys():
            word_to_cnt[word3] += 1
        else:
            word_to_cnt[word3] = 1
        if labels[i] >= pos_threshold:
            if word3 in word3_to_positive.keys():
                word3_to_positive[word3] += 1
            else:
                word3_to_positive[word3] = 1
        if labels[i] <= neg_threshold:
            if word3 in word3_to_negative.keys():
                word3_to_negative[word3] += 1
            else:
                word3_to_negative[word3] = 1
                
    for word in word_to_cnt.keys():
        word3_line_to_cnt[(word, i)] = word_to_cnt[word]
```

```python
for word in word_to_index3.keys():
    if word not in word_to_positive3.keys():
        word3_to_positive[word] = 0.00000000000000000001
    if word not in word_to_negative.keys():
        word3_to_negative[word] = 0.00000000000000000001   
```

```python
features = []
for i in range(len(lines)):
    line = lines[i]
    line2 = lines[i]
    line3 = lines_3[i]
    feature = np.zeros(len(words) + len(words2) + len(words3))
    for word in line:
        if word in word_to_index.keys():
            feature[word_to_index[word]] = word_line_to_cnt[(word, i)] * np.log(negative_number * word_to_positive[word] / positive_number / word_to_negative[word])
    for word2 in line2:
        if word2 in word_to_index2.keys():
            feature[len(words) + word_to_index2[word2]] = 1
        if word3 in word_to_index3.keys():
            feature[len(words) + len(words2) + word_to_index3[word3]] = 1
    features.append(feature)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor()
```

```python
clf.fit(list(features), list(labels))  
```
